```python
from gnews import GNews
import pandas as pd
from datetime import datetime

# Cấu hình GNews
google_news = GNews(language='vi', country='Vietnam',
                    start_date=(2024, 1, 1), end_date=(2025, 6, 1), max_results=20)

# Nhóm từ khóa
keyword_groups = {
    "vi_mo": ["GDP Việt Nam", "CPI", "lạm phát", "chính sách tiền tệ",
        "chính sách tài khóa", "tăng trưởng kinh tế", "tỷ giá", "nợ công"],  # dán từ khóa ở đây
    "ngan_hang": ["lãi suất ngân hàng", "Ngân hàng Nhà nước", "hạn mức tín dụng",
        "nợ xấu", "vốn hóa ngân hàng", "chính sách tín dụng", "thanh khoản"],
    "chung_khoan": ["VN-Index", "thị trường chứng khoán", "phái sinh", "room ngoại",
        "ETF", "giá cổ phiếu", "tăng trần", "khối ngoại mua ròng"]
}

# Kết quả tổng hợp
all_articles = []

# Lặp từng nhóm từ khóa
for group, keywords in keyword_groups.items():
    for kw in keywords:
        try:
            articles = google_news.get_news(kw)
            for article in articles:
                all_articles.append({
                    "keyword": kw,
                    "group": group,
                    "title": article.get('title'),
                    "summary": article.get('description', ''),  # Lấy mô tả nếu có
                    "date": article.get('published date'),
                    "link": article.get('url'),
                    "source": article.get("publisher", "Unknown")
                })
        except Exception as e:
            print(f"⚠️ Lỗi với từ khóa '{kw}': {e}")

# Xuất ra CSV
df = pd.DataFrame(all_articles)
df['date'] = pd.to_datetime(df['date']).dt.date  # chuẩn hóa ngày
df = df.drop_duplicates(subset=['title', 'link'])  # loại trùng

filename = f"tin_tuc_vimo_nganh_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
df.to_csv(filename, index=False, encoding='utf-8-sig')
print(f"✅ Đã lưu {len(df)} tin vào {filename}")
```

```python
import pandas as pd
from transformers import pipeline
from googletrans import Translator
from tqdm import tqdm

# Load file CSV đã có
df = pd.read_csv("tin_tuc_vimo_nganh_20250728_004905.csv")  # Thay bằng tên file thật

# Thiết lập mô hình sentiment tiếng Anh
classifier = pipeline("sentiment-analysis", model="j-hartmann/emotion-english-distilroberta-base")
translator = Translator()

# Thêm cột sentiment và score
sentiments = []
scores = []

for _, row in tqdm(df.iterrows(), total=len(df)):
    try:
        # Kết hợp tiêu đề + mô tả (nếu có)
        text_vi = row['title']
        if 'summary' in df.columns:
            text_vi += " " + str(row['summary'])

        # Dịch sang tiếng Anh
        text_en = translator.translate(text_vi, src='vi', dest='en').text

        # Phân tích cảm xúc
        result = classifier(text_en)[0]
        sentiments.append(result['label'])
        scores.append(round(result['score'], 3))

    except Exception as e:
        sentiments.append("neutral")
        scores.append(0.5)
        print(f"⚠️ Lỗi xử lý dòng: {e}")

# Thêm vào DataFrame
df['sentiment'] = sentiments
df['score'] = scores

# Lưu lại
output_file = "tin_tuc_phan_tich_cam_xuc.csv"
df.to_csv(output_file, index=False, encoding='utf-8-sig')
print(f"✅ Đã lưu file mới với phân tích sentiment: {output_file}")
```

```python
import pandas as pd

# Tạo dải ngày từ 01-01-2024 đến 01-06-2025
date_range = pd.date_range(start="2024-01-01", end="2025-06-01")

# Tạo bảng dim_date
dim_date = pd.DataFrame({
    "full_date": date_range,
})

dim_date["date_key"] = dim_date["full_date"].dt.strftime('%Y%m%d').astype(int)
dim_date["year"] = dim_date["full_date"].dt.year
dim_date["quarter"] = dim_date["full_date"].dt.quarter
dim_date["month"] = dim_date["full_date"].dt.month
dim_date["week"] = dim_date["full_date"].dt.isocalendar().week
dim_date["day"] = dim_date["full_date"].dt.day
dim_date["day_name"] = dim_date["full_date"].dt.day_name()
dim_date["is_weekend"] = dim_date["day_name"].isin(["Saturday", "Sunday"])

# Sắp xếp theo ngày
dim_date = dim_date.sort_values("full_date")

# Xuất ra file nếu cần
dim_date.to_excel("dim_date.xlsx", index=False)
print("✅ Đã tạo bảng dim_date.xlsx")
```

```python
import pandas as pd

# === 1. Đọc dữ liệu từ 3 file Excel gốc ===
gia_chung_khoan_df = pd.read_excel("gia_chung_khoan_gop.xlsx")
gia_ngan_hang_df = pd.read_excel("gia_nganh_ngan_hang_gop.xlsx")
tin_tuc_df = pd.read_excel("tin_tuc_phan_tich_cam_xuc.xlsx")

# === 2. Gộp dữ liệu giá cổ phiếu ===
gia_df = pd.concat([gia_chung_khoan_df, gia_ngan_hang_df], ignore_index=True)
gia_df["StockID"] = gia_df["nganh"]
gia_df["DateKey"] = gia_df["time"].dt.date

# === 3. Tạo Dim_Stock ===
nganh_ngan_hang = gia_ngan_hang_df["nganh"].unique().tolist()
dim_stock = gia_df[["StockID", "nganh"]].drop_duplicates().rename(columns={"nganh": "Industry"})
dim_stock["IndustryGroup"] = dim_stock["StockID"].apply(
    lambda x: "Ngân hàng" if x in nganh_ngan_hang else "Tài chính"
)

# === 4. Tạo Dim_Date ===
dim_date = pd.DataFrame(gia_df["DateKey"].drop_duplicates())
dim_date["Day"] = dim_date["DateKey"].apply(lambda x: x.day)
dim_date["Month"] = dim_date["DateKey"].apply(lambda x: x.month)
dim_date["Quarter"] = dim_date["DateKey"].apply(lambda x: (x.month - 1) // 3 + 1)
dim_date["Year"] = dim_date["DateKey"].apply(lambda x: x.year)
dim_date["Weekday"] = dim_date["DateKey"].apply(lambda x: x.strftime("%A"))

# === 5. Xử lý bảng tin tức ===
tin_tuc_df["DateKey"] = pd.to_datetime(tin_tuc_df["date"]).dt.date

dim_news = tin_tuc_df.copy()
dim_news["NewsID"] = range(1, len(dim_news) + 1)
dim_news = dim_news[[
    "NewsID", "title", "summary", "keyword", "group",
    "sentiment", "score", "DateKey"
]].rename(columns={
    "title": "Title",
    "summary": "Summary",
    "keyword": "Keyword",
    "group": "Group",
    "sentiment": "Sentiment",
    "score": "Score"
})

# === 6. Tổng hợp cảm xúc theo ngày ===

def dominant_sentiment(series):
    return series.value_counts().idxmax()

# Tìm sentiment chính mỗi ngày
dominant_df = tin_tuc_df.groupby("DateKey")["sentiment"].agg(dominant_sentiment).reset_index()
dominant_df = dominant_df.rename(columns={"sentiment": "SentimentLevel"})

# Gắn cảm xúc chính vào từng dòng để lọc score phù hợp
tin_tuc_with_dominant = tin_tuc_df.merge(dominant_df, on="DateKey", how="left")
tin_tuc_with_dominant["match"] = tin_tuc_with_dominant["sentiment"] == tin_tuc_with_dominant["SentimentLevel"]

# Tính trung bình score của những tin có sentiment khớp sentiment chính
sentiment_score_df = tin_tuc_with_dominant[tin_tuc_with_dominant["match"]].groupby("DateKey")["score"].mean().reset_index()
sentiment_score_df = sentiment_score_df.rename(columns={"score": "SentimentScore"})

# Đếm số lượng tin trong ngày
news_count_df = tin_tuc_df.groupby("DateKey")["title"].count().reset_index().rename(columns={"title": "NewsCount"})

# Gộp lại thành bảng tổng hợp cảm xúc theo ngày
sentiment_summary = dominant_df.merge(sentiment_score_df, on="DateKey", how="left")
sentiment_summary = sentiment_summary.merge(news_count_df, on="DateKey", how="left")

# === 7. Tạo bảng Fact ===
fact_df = gia_df.merge(sentiment_summary, on="DateKey", how="left")
fact_stock_price_news = fact_df[[
    "DateKey", "StockID", "open", "high", "low", "close", "volume",
    "SentimentScore", "SentimentLevel", "NewsCount"
]].rename(columns={
    "open": "OpenPrice",
    "high": "HighPrice",
    "low": "LowPrice",
    "close": "ClosePrice",
    "volume": "Volume"
})

# === 8. Ghi ra file Excel ===
output_file = "Stock_News_Model_WithSentimentScore.xlsx"
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    fact_stock_price_news.to_excel(writer, sheet_name="Fact_StockPriceNewsImpact", index=False)
    dim_date.to_excel(writer, sheet_name="Dim_Date", index=False)
    dim_stock.to_excel(writer, sheet_name="Dim_Stock", index=False)
    dim_news.to_excel(writer, sheet_name="Dim_News", index=False)

print(f"✅ File '{output_file}' đã được tạo thành công.")
```
